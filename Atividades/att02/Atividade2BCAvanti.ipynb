1. Escreva uma função que receba uma lista de números e retorne outra lista
com os números ímpares.

Para encontrar os números ímpares de uma lista, percorremos cada elemento e verificamos o resto da divisão por 2 usando o operador %. Se o resto for diferente de zero, o número é ímpar e é adicionado a uma nova lista. Essa abordagem tem complexidade O(n), pois passa uma única vez por todos os elementos.


```python
lista=[1,2,3,4,5,6,7]

def encontrar_impares(lista):
    impares = []
    for numero in lista:
        if numero % 2 != 0:
            impares.append(numero)
    return impares

print(encontrar_impares(lista))
```

2. Escreva uma função que receba uma lista de números e retorne outra lista
com os números primos presentes.

Um número é primo quando é maior que 1 e divisível apenas por 1 e por ele mesmo. Na função, para cada número da lista testamos se ele é divisível por algum valor entre 2 e (número - 1). Se encontrarmos um divisor, o break interrompe o laço e o número é descartado. Se nenhum divisor for encontrado, o else do for é executado e o número é considerado primo. Essa técnica de "divisão por tentativa" é simples de entender, embora para listas com números muito grandes seja possível otimizar testando divisores apenas até a raiz quadrada do número.

```python
lista=[1,2,3,4,5,6,7]
def encontrar_primos(lista):
    primos = []
    for numero in lista:
        if numero > 1:
            for i in range(2, numero):
                if (numero % i) == 0:
                    break
            else:
                primos.append(numero)
    return primos

print(encontrar_primos(lista))
```

3. Escreva uma função que receba duas listas e retorne outra lista com os
elementos que estão presentes em apenas uma das listas.
Para obter os elementos que aparecem em apenas uma das duas listas (e não nas duas ao mesmo tempo), usamos a operação de diferença simétrica entre conjuntos (^). Diferente da união (que traria todos os elementos) ou da interseção (que traria só os elementos comuns), a diferença simétrica retorna exatamente os elementos que existem em um conjunto ou no outro, mas não em ambos. Convertemos as listas em set para poder aplicar essa operação e depois voltamos para list no resultado final.

```python
def elementos_exclusivos(lista1, lista2):
    exclusivos = set(lista1) ^ set(lista2)
    return list(exclusivos)

A = [1, 2, 3, 4]
B = [3, 4, 5, 6]
print(elementos_exclusivos(A, B))
```

4. Dada uma lista de números inteiros, escreva uma função para encontrar o
segundo maior valor na lista.
A função primeiro identifica o maior valor da lista com max(). Em seguida, percorre a lista novamente ignorando ocorrências desse valor máximo e vai guardando o maior número encontrado entre os restantes — esse é o "segundo maior".

Vale observar uma escolha de interpretação: se houver valores repetidos no topo (por exemplo, [7, 7, 3]), a função trata o segundo maior como o segundo valor distinto, retornando 3 em vez de 7. Essa é uma forma comum de interpretar o problema, mas dependendo do contexto poderia-se querer o segundo elemento em ordem de grandeza, incluindo repetições — o que exigiria uma abordagem por ordenação (sorted) ao invés de comparação direta com o máximo.

```python
lista=[1,2,3,4,5,6,7]

def encontrar_segundo_maior(lista):
    maior =max(lista)
    segundo_maior=None
    for numero in lista:
      if numero != maior:
        if segundo_maior is None or numero > segundo_maior:
          segundo_maior = numero
    return segundo_maior

print(encontrar_segundo_maior(lista))
```

5. Crie uma função que receba uma lista de tuplas, cada uma contendo o
nome e a idade de uma pessoa, e retorne a lista ordenada pelo nome das
pessoas em ordem alfabética.

Para ordenar a lista de tuplas pelo nome, usamos a função sorted() com o parâmetro key=lambda x: x[0], que indica que a comparação entre os elementos deve ser feita usando a primeira posição de cada tupla (o nome). Por padrão, sorted() já ordena strings em ordem alfabética (ordem lexicográfica), então não é necessário nenhum tratamento adicional.

```python
def ordenar_pessoas(pessoas):
  minha_tupla=tuple(pessoas)
  lista_ordenada = sorted(minha_tupla, key=lambda x: x[0])
  return lista_ordenada

pessoas = [("Charlie", 20), ("Alice", 25), ("Bob", 30)]
print(ordenar_pessoas(pessoas))
```

6. Como identificar e tratar outliers em uma coluna numérica usando desvio
padrão ou quartis?

Outliers são valores que se distanciam muito do padrão geral dos dados e podem distorcer análises estatísticas. Duas abordagens comuns para identificá-los são:

Método do IQR (intervalo interquartil): calculamos o primeiro quartil (Q1, 25%) e o terceiro quartil (Q3, 75%). O IQR é a diferença Q3 − Q1. Qualquer valor abaixo de Q1 − 1.5×IQR ou acima de Q3 + 1.5×IQR é considerado outlier. É um método robusto, pouco sensível a valores extremos.
Método do desvio padrão (Z-score): calculamos a média e o desvio padrão da coluna. Valores que ficam a mais de 2 ou 3 desvios padrão de distância da média são considerados outliers. Esse método assume que os dados seguem aproximadamente uma distribuição normal.

Depois de identificados, os outliers podem ser removidos (excluindo as linhas) ou tratados por Winsorization, técnica em que os valores extremos são "cortados" (clip) para os limites inferior/superior calculados, em vez de descartados — preservando o tamanho da amostra.

No exemplo, o valor 100 é claramente um outlier: ambos os métodos o identificam corretamente, mas o limite do desvio padrão ficou bem mais largo (por ser sensível ao próprio valor extremo, que "puxa" a média e o desvio para cima), o que ilustra por que o IQR costuma ser preferido quando há valores muito discrepantes.

```python
import pandas as pd
import numpy as np

# Sample DataFrame for demonstration
data = {'coluna': [10, 12, 12, 13, 12, 11, 14, 13, 15, 10, 12, 100, 11, 16, 12, 17]}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# --- Identificação e tratamento de outliers usando Quartis (IQR) ---
Q1 = df['coluna'].quantile(0.25)
Q3 = df['coluna'].quantile(0.75)
IQR = Q3 - Q1

limite_inferior = Q1 - (1.5 * IQR)
limite_superior = Q3 + (1.5 * IQR)

print(f"\nLimite Inferior (IQR): {limite_inferior}")
print(f"Limite Superior (IQR): {limite_superior}")

# Outliers identificados
outliers_iqr = df[(df['coluna'] < limite_inferior) | (df['coluna'] > limite_superior)]
print("\nOutliers identificados (IQR):")
print(outliers_iqr)

# Remover outliers
df_sem_outliers_iqr = df[(df['coluna'] >= limite_inferior) & (df['coluna'] <= limite_superior)]
print("\nDataFrame sem outliers (IQR):")
print(df_sem_outliers_iqr)

# OU Substituir pelo limite (Winsorization com IQR)
df['coluna_tratada_iqr'] = df['coluna'].clip(lower=limite_inferior, upper=limite_superior)
print("\nDataFrame com outliers tratados (Winsorization IQR):")
print(df)

# --- Identificação e tratamento de outliers usando Desvio Padrão (Z-score) ---
media = df['coluna'].mean()
desvio_padrao = df['coluna'].std()

# Geralmente, valores fora de 2 ou 3 desvios padrão são considerados outliers
num_desvios = 2 # Pode ser ajustado

limite_inferior_sd = media - (num_desvios * desvio_padrao)
limite_superior_sd = media + (num_desvios * desvio_padrao)

print(f"\nLimite Inferior ({num_desvios} Desvios Padrão): {limite_inferior_sd:.2f}")
print(f"Limite Superior ({num_desvios} Desvios Padrão): {limite_superior_sd:.2f}")

# Outliers identificados
outliers_sd = df[(df['coluna'] < limite_inferior_sd) | (df['coluna'] > limite_superior_sd)]
print("\nOutliers identificados (Desvio Padrão):")
print(outliers_sd)

# Remover outliers
df_sem_outliers_sd = df[(df['coluna'] >= limite_inferior_sd) & (df['coluna'] <= limite_superior_sd)]
print("\nDataFrame sem outliers (Desvio Padrão):")
print(df_sem_outliers_sd)

# OU Substituir pelo limite (Winsorization com Desvio Padrão)
df['coluna_tratada_sd'] = df['coluna'].clip(lower=limite_inferior_sd, upper=limite_superior_sd)
print("\nDataFrame com outliers tratados (Winsorization Desvio Padrão):")
print(df)
```

7. Como concatenar vários DataFrames (empilhando linhas ou colunas),
mesmo que tenham colunas diferentes?
Dica: Utiliza-se pd.concat() especificando axis=0 (linhas) ou axis=1
(colunas). Quando há colunas diferentes, os valores ausentes são
preenchidos com NaN.

O pd.concat() permite empilhar DataFrames tanto por linhas (axis=0) quanto por colunas (axis=1).


axis=0: os DataFrames são empilhados um embaixo do outro. Se as colunas forem diferentes entre eles, o pandas alinha pelo nome da coluna e preenche com NaN onde não houver correspondência (por exemplo, um DataFrame sem a coluna "Cidade" recebe NaN nessa coluna ao ser empilhado com outro que a possui).
axis=1: os DataFrames são colocados lado a lado. Nesse caso, o pandas alinha pelo índice das linhas, não pelo conteúdo. Por isso, é importante garantir que os índices representem a mesma entidade nos dois DataFrames (por exemplo, usando set_index('ID')) antes de concatenar, senão o resultado pode juntar linhas que não correspondem à mesma pessoa/registro.

```python
import pandas as pd

# Criando o primeiro DataFrame (df1)
data1 = {'ID': [1, 2, 3],
         'Nome': ['Alice', 'Bob', 'Charlie'],
         'Idade': [25, 30, 35]}
df1 = pd.DataFrame(data1)
print("DataFrame 1:")
print(df1)
print("\n" + "-"*30 + "\n")

# Criando o segundo DataFrame (df2) com algumas colunas diferentes
data2 = {'ID': [4, 5],
         'Nome': ['David', 'Eve'],
         'Cidade': ['Nova Iorque', 'Londres']}
df2 = pd.DataFrame(data2)
print("DataFrame 2:")
print(df2)
print("\n" + "-"*30 + "\n")

# --- Concatenação por linhas (axis=0) ---
# Empilhando df1 e df2. Observe as colunas 'Idade' e 'Cidade' e os NaNs.
print("Concatenando por linhas (axis=0):")
df_linhas = pd.concat([df1, df2], axis=0)
print(df_linhas)
print("\n" + "-"*30 + "\n")

# Criando um terceiro DataFrame (df3) com outras colunas para concatenação por colunas
data3 = {'Produto': ['Maçã', 'Banana', 'Laranja'],
         'Preço': [1.50, 0.75, 2.00]}
df3 = pd.DataFrame(data3)
print("DataFrame 3:")
print(df3)
print("\n" + "-"*30 + "\n")

# --- Concatenação por colunas (axis=1) ---
# Colocando df1 e df3 lado a lado. Observe como as linhas se alinham pelos índices.
# Como os índices não são os mesmos para todos os dados, teremos NaNs.
print("Concatenando por colunas (axis=1):")
df_colunas = pd.concat([df1, df3], axis=1)
print(df_colunas)
print("\n" + "-"*30 + "\n")

# Exemplo mais comum de axis=1 quando os DataFrames têm as mesmas linhas
# mas queremos adicionar novas colunas de outro DataFrame
data4 = {'ID': [1, 2, 3],
         'Salario': [50000, 60000, 75000]}
df4 = pd.DataFrame(data4)

print("DataFrame 4:")
print(df4)
print("\n" + "-"*30 + "\n")

print("Concatenando por colunas (axis=1) com mesmos IDs (mais comum para juntar informações):")
df_colunas_com_id = pd.concat([df1, df4.set_index('ID')], axis=1, join='inner') # Usamos set_index para alinhar por ID
print(df_colunas_com_id)
```

9. Utilizando pandas, como selecionar uma coluna específica e filtrar linhas
em um “DataFrame” com base em uma condição?

Existem duas formas equivalentes de filtrar linhas de um DataFrame com base em uma condição e selecionar uma coluna específica:


Notação de colchetes: df[df['Idade'] > 25]['Nome'] — primeiro criamos uma máscara booleana (df['Idade'] > 25), que é usada para filtrar as linhas, e depois selecionamos a coluna desejada.
.loc[]: df.loc[df['Idade'] > 25, 'Nome'] — faz a mesma coisa, mas em uma única operação, especificando a condição de linhas e a coluna desejada juntas. É a forma recomendada pela documentação do pandas, pois evita o SettingWithCopyWarning que pode aparecer ao encadear operações com colchetes.

```python
import pandas as pd

# Exemplo de DataFrame
dados = {
    'Nome': ['Ana', 'Bruno', 'Carlos'],
    'Idade': [22, 35, 28],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'São Paulo']
}
df = pd.DataFrame(dados)

# 1. Filtrar linhas onde a Idade é maior que 25 e selecionar a coluna 'Nome'
df_filtrado = df[df['Idade'] > 25]['Nome']
print(df_filtrado)

# 2. Utilizando o método .loc para o mesmo resultado (ideal para evitar avisos futuros)
df_filtrado_loc = df.loc[df['Idade'] > 25, 'Nome']
print(df_filtrado_loc)
```

10. Utilizando pandas, como lidar com valores ausentes (NaN) em um
DataFrame?

10. Lidar com valores ausentes (NaN)

O pandas oferece diferentes estratégias para lidar com dados ausentes:


Identificação: df.isna() retorna um DataFrame booleano indicando onde há valores ausentes; combinado com .sum(), mostra quantos NaN existem por coluna.
Remoção (dropna()): remove linhas (ou colunas, com axis=1) que contenham pelo menos um valor ausente. É uma solução simples, mas pode descartar informação valiosa se muitas linhas tiverem algum NaN.
Preenchimento (fillna()): substitui os valores ausentes por algo definido — pode ser um valor fixo, um texto padrão, ou uma estatística como a média/mediana da própria coluna. É preferível à remoção quando não queremos perder linhas inteiras, mas exige cuidado para não introduzir viés nos dados (por exemplo, preencher com a média pode reduzir artificialmente a variância da coluna).

A escolha entre remover ou preencher depende do contexto: quantos dados estão faltando, se o motivo da ausência é aleatório e qual o impacto de cada abordagem na análise final.

```python
import numpy as np
import pandas as pd

# 1. Criando um DataFrame de exemplo com valores ausentes (NaN)
dados = {
    "Nome": ["Ana", "Bruno", "Carlos", "Diana"],
    "Idade": [25, np.nan, 30, 22],
    "Salario": [5000, 4000, np.nan, 4500],
    "Cidade": ["São Paulo", "Rio", "Belo Horizonte", None],
}
df = pd.DataFrame(dados)
print("--- DataFrame Original ---")
print(df)

# 2. Identificando os valores ausentes
print("\n--- Quantidade de NaN por coluna ---")
print(df.isna().sum())

# 3. Exemplo de Remoção (.dropna)
# Remove a linha do Carlos (Salario NaN) e da Diana (Cidade None)
df_sem_na = df.dropna()
print("\n--- Linhas com qualquer NaN removidas ---")
print(df_sem_na)

# 4. Exemplo de Preenchimento (.fillna)
df_preenchido = df.copy()

# Preenche a Idade com a média das idades (25.66)
media_idade = df_preenchido["Idade"].mean()
df_preenchido["Idade"] = df_preenchido["Idade"].fillna(media_idade)

# Preenche o Salário com um valor fixo (0)
df_preenchido["Salario"] = df_preenchido["Salario"].fillna(0)

# Preenche a Cidade com um texto padrão
df_preenchido["Cidade"] = df_preenchido["Cidade"].fillna("Não Informado")

print("\n--- DataFrame com valores preenchidos ---")
print(df_preenchido)
```
